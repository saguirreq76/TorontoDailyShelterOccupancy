```python
# 2017
SectorShelter2017 = Shelter2017.groupby(['FACILITY_NAME', 'SECTOR'])['SECTOR'].count().unstack(fill_value=0)

# 2018
SectorShelter2018 = Shelter2018.groupby(['FACILITY_NAME', 'SECTOR'])['SECTOR'].count().unstack(fill_value=0)

# 2019
SectorShelter2019 = Shelter2019.groupby(['FACILITY_NAME', 'SECTOR'])['SECTOR'].count().unstack(fill_value=0)

# 2020
SectorShelter2020 = Shelter2020.groupby(['FACILITY_NAME', 'SECTOR'])['SECTOR'].count().unstack(fill_value=0)

Sector_NonZeroCount = {
    2017: (SectorShelter2017 != 0).astype(int).sum(),
    2018: (SectorShelter2018 != 0).astype(int).sum(),
    2019: (SectorShelter2019 != 0).astype(int).sum(),
    2020: (SectorShelter2020 != 0).astype(int).sum()
}

Sector_NonZeroCount_df = pd.DataFrame(Sector_NonZeroCount)

colors = ["#57A8B2", "#23757F", "#D1CDE5", "#E8B3C9", "#B26A91"]

plt.figure(figsize=(6, 6))
ax = Sector_NonZeroCount_df.T.plot(kind='bar', color=colors)
plt.title('Count of Shelters by Sector', fontweight="bold", fontsize=14)
plt.xlabel('Year', fontweight="bold", fontsize=12)
plt.ylabel('')
plt.xticks(rotation=0)
plt.legend(title='Sector', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# 2017
DiffShelter2017 = Shelter2017.copy()
DiffShelter2017["DIFF"]= Shelter2017.CAPACITY - Shelter2017.OCCUPANCY

# 2018
DiffShelter2018 = Shelter2018.copy()
DiffShelter2018["DIFF"]= Shelter2018.CAPACITY - Shelter2018.OCCUPANCY

# 2019
DiffShelter2019 = Shelter2019.copy()
DiffShelter2019["DIFF"]= Shelter2019.CAPACITY - Shelter2019.OCCUPANCY

# 2020
DiffShelter2020 = Shelter2020.copy()
DiffShelter2020["DIFF"]= Shelter2020.CAPACITY - Shelter2020.OCCUPANCY

DiffShelterSum2017 = DiffShelter2017["DIFF"].sum()
DiffShelterSum2018 = DiffShelter2018["DIFF"].sum()
DiffShelterSum2019 = DiffShelter2019["DIFF"].sum()
DiffShelterSum2020 = DiffShelter2020["DIFF"].sum()

YearDiff = {2017: DiffShelterSum2017, 2018:DiffShelterSum2018, 2019: DiffShelterSum2019, 2020: DiffShelterSum2020}
YearDiff_df = pd.DataFrame.from_dict(YearDiff, orient='index', columns=['DifferenceSum'])

colors = ["#57A8B2", "#23757F", "#D1CDE5", "#E8B3C9"]

plt.figure(figsize=(6, 6))
bars = plt.bar(YearDiff_df.index, YearDiff_df["DifferenceSum"], color=colors)
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('')
plt.tick_params(axis='y', which='both', length=0)
plt.gca().set_yticklabels([])
plt.title('Unused Portion by Year', fontsize=14, fontweight='bold')
plt.xticks(YearDiff_df.index)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{int(yval)}', ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020]

DiffShelters = {}

for year in years:
    Shelters = globals()[f"Shelter{year}"]
    DiffShelter = Shelters.copy()
    DiffShelter["DIFF"] = Shelters["CAPACITY"] - Shelters["OCCUPANCY"]
    DiffShelter["MONTH"] = pd.to_datetime(Shelters["OCCUPANCY_DATE"]).dt.month
    DiffShelters[year] = DiffShelter

colors = ["#57A8B2", "#23757F", "#D1CDE5", "#E8B3C9"]
width = 0.2

plt.figure(figsize=(10, 6))

diff_sum_by_month_all_years = pd.DataFrame()

for i, year in enumerate(years):
    diff_shelter = DiffShelters[year]
    diff_sum_by_month = diff_shelter.groupby('MONTH')['DIFF'].sum()
    diff_sum_by_month_all_years[year] = diff_sum_by_month

    plt.bar(diff_sum_by_month.index + i * width, diff_sum_by_month, width=width, label=str(year), color=colors[i])

plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Unused Count', fontsize=12, fontweight='bold')
plt.title('Unused Portion by Month', fontsize=14, fontweight='bold')
plt.xticks(diff_sum_by_month.index + width * (len(years) - 1) / 2, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()

```
